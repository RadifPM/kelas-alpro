```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('style/sci.mplstyle')
import numpy as np

# Open and read the file gr.dos
x, y = np.loadtxt('baseball.txt', unpack = True)
    
# Create figure object
plt.figure()
# Plot the DOS, in which the Fermi energy shifts to zero
plt.plot(x, y, color='red')
#coba
plt.title('Trajectory of a batted baseball')
# Add the x and y-axis labels
plt.xlabel('x(m)')
plt.ylabel('y(m)')
# Set the axis limits
plt.xlim(0, 150)
plt.ylim(0, 35)
# Save the figure to the pdf file
plt.savefig('plot-dos.pdf', bbox_inches='tight')
```
